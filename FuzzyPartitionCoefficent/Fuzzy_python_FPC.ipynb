## Imagen Blanco y Negro

### Corte Axial

```python
im = Image.open('/home/isa_ubuntu_16_04/Escritorio/FASE_6_DL/clustering_Fuzz/3.png')
im2arr = np.array(im) # im2arr.shape: height x width x channel
arr2im = Image.fromarray(im2arr)
imgplot = plt.imshow(arr2im)
```

```python
img = imread('/home/isa_ubuntu_16_04/Escritorio/FASE_6_DL/clustering_Fuzz/3.png')
print(type(img))
img.shape
```

```python
print(img)
```

```python
img_2=img [0:1]
print(img_2)
print(img_2.shape)
```

```python
fig1, axes1 = plt.subplots(3, 3, figsize=(10, 10))

fpcs = []
colors = ['b', 'orange', 'g', 'r', 'c', 'm', 'y', 'k', 'Brown', 'ForestGreen']

for ncenters, ax in enumerate(axes1.reshape(-1), 2):
    cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
        img, ncenters, 2, error=0.005, maxiter=1000, init=None)

    # Store fpc values for later
    fpcs.append(fpc)

    # Plot assigned clusters, for each data point in training set
    cluster_membership = np.argmax(u, axis=0)
    for j in range(ncenters):
        ax.plot(img [0][cluster_membership == j],
                img [1][cluster_membership == j], '.', color=colors[j])

    # Mark the center of each fuzzy cluster
    for pt in cntr:
        ax.plot(pt[0], pt[1], 'rs')

    ax.set_title('Centers = {0}; FPC = {1:.2f}'.format(ncenters, fpc), size=12)
    ax.axis('off')

fig1.tight_layout()
```

```python
fig2, ax2 = plt.subplots()
ax2.plot(np.r_[2:11], fpcs, color='#e809d5')
ax2.set_title("How Number of Clusters Change FPC?")
ax2.set_xlabel("Number of centers")
ax2.set_ylabel("Fuzzy partition coefficient")
plt.show()
```
